```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.utils import shuffle
```

```python
T = 100
x_axis = np.linspace(0, 2*np.pi, T)
y_axis = np.sin(x_axis)
```

```python
N = 30
idx = np.random.choice(T, size = N, replace = False)
idx
```

```python
X_train = x_axis[idx].reshape(N, 1)
y_train = y_axis[idx]
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)

predictions = model.predict(x_axis.reshape(T,1))

print("model score : " + str(model.score(x_axis.reshape(T,1), y_axis)))
```

```python
plt.plot(x_axis, y_axis, color = "red", label = "actual sin(x)")
plt.plot(x_axis, predictions, color = "blue", label = "dec tree regressor")
plt.legend()
plt.show()
```

```python
class BaggedTreeRegressor:
    
    def __init__(self, B):
        self.B = B
        
    def fit(self, X, y):
        self.models = []
        
        for b in range(0, self.B):
            idx = np.random.choice(N, size = N, replace = True)
            Xb = X[idx].reshape(N, 1)
            yb = y[idx]
            
            model = DecisionTreeRegressor()
            model.fit(Xb, yb)
            
            self.models.append(model)
    
    def predict(self, X):
        predictions = np.zeros(len(X))
        
        for model in self.models:
            predictions += model.predict(X.reshape(len(X),1))
        
        return predictions/self.B
    
    def score(self, X, Y):
        d1 = Y - self.predict(X)
        d2 = Y - Y.mean()
        
        return 1 - d1.dot(d1)/d2.dot(d2)
    
```

```python
model_bagged = BaggedTreeRegressor(B = 200)
model_bagged.fit(X_train, y_train)

print("model_bagged score : " + str(model_bagged.score(x_axis.reshape(len(x_axis),1), y_axis)))
```

```python
bagged_predictions = model_bagged.predict(x_axis)
len(bagged_predictions)
```

```python
plt.plot(x_axis, y_axis, color = "red", label = "actual sin(x)")
plt.plot(x_axis, predictions, color = "blue", label = "dec tree regressor")
plt.legend()
plt.show()
```

```python
plt.clf()
plt.plot(x_axis, y_axis, color = "red", label = "actual sin(x)")
plt.plot(x_axis, bagged_predictions, color = "green", label = "bagged dec tree regressor")
plt.legend()
plt.show()
```
